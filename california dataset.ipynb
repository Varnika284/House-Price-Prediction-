```python
#Handle missing value
#drop a particular column
data_na=data.dropna(subset=["AveBedrms"])
data_na.shape
```

```python
#categorical variable by Creating a dummy variable- binary variable
dummies=pd.get_dummies(data_na.Population)
dummies
```

```python
data_na_dummies=pd.concat([data_na, dummies], axis='columns')
data_na_dummies
```

```python
data_na_dummies.head()
```

```python
#To drop the column
data_clean=data_na_dummies.drop(['AveBedrms', 12873.0], axis='columns')
data_clean.head()
```

```python
#creating Features and labeled datasets
X=data_clean.drop(columns=['AveRooms'])
X.head()
```

```python
y=data_clean['AveRooms']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1984)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
OLS = LinearRegression()

OLS.fit(X_train,y_train)
```

```python
#display the intercept and coefficients of the OLS model
print ("Intetrcept is " + str(OLS.intercept_))
print ("The set of coefficients are " + str(OLS.coef_))
print ("The R-Squared value is " + str(OLS.score(X_train,y_train)))
```

```python
#predicting with OLS
y_pred=OLS.predict(X_test)
performance=pd.DataFrame({'PREDICTIONS':y_pred,'ACTUAL VALUES':y_test})
performance['error']=performance['ACTUAL VALUES']-performance['PREDICTIONS']
performance.head()
```

```python
#preparing data for plotting

performance.reset_index(drop=True,inplace=True)
performance.reset_index(inplace=True)
performance.head()
```

```python
#plot the residuals 

fig = plt.figure(figsize=(10,5))
plt.bar('index', 'error', data=performance[:50], color='black', width=0.3)
plt.xlabel("Observations")
plt.ylabel("residuals")
plt.show()
```
